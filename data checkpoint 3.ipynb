```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220, 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .95 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(data)
```

this function returns a correlation heatmap with several Parameters:
-figsize: sets the distance between the heatmap and the colorbar and the volume and length of the colorbar
-sns.diverging_palette(): sets the color range of the heatmap
-square:if True, sets the Axes aspect to “equal” so each cell will be square-shaped.
-cbar_kws: sets the length of the colorbar
-annote :shows the values on the squares
-annot_kws:sets the size of values when annot is True
